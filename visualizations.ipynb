```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set the data
labels = ["Fixed", "Heuristic", "RL-Model"]
all_good =  [486.1616, 440.8526, 382.576]
one_light_down = [482.7623, 444.5843, 393.2356]
two_light_down = [482.5383, 445.817, 398.5483]
```

```python
%matplotlib inline
```

```python
# Set the width of the bars
bar_width = 0.25

# Set the x positions of the bars
r1 = np.arange(len(all_good))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bars
plt.bar(r1, all_good, color='b', width=bar_width, edgecolor='grey', label='Normal')
plt.bar(r2, one_light_down, color='g', width=bar_width, edgecolor='grey', label='1 light down')
plt.bar(r3, two_light_down, color='r', width=bar_width, edgecolor='grey', label='2 lights down')

# Add labels, title, and legend
plt.xlabel("Models")
plt.ylabel("Average Car Travel Time (seconds)")
plt.title("Fig. 1 - Traffic Control Models Comparison")
plt.xticks([r + bar_width for r in range(len(all_good))], labels)
plt.ylim(300, 500)
plt.legend(loc='lower right')

# Remove right and top axis spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Set a gray background
plt.gca().set_facecolor('lightgrey')

# Add a grid
plt.grid(axis='y', color='white', linestyle='-', linewidth=0.5)

# Set a less abrasive color for the bars
plt.gca().set_prop_cycle('color', ['#6EB5FF', '#9AFF9A', '#FF9999'])
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the data
labels = ["Normal", "1 light down", "2 lights down"]
fixed =  [486.1616, 482.7623, 482.5383]
heuristic = [451.8526, 444.5843, 445.817]
ml_model = [382.576, 393.2356, 398.5483]

# Set the width of the bars
bar_width = 0.25

# Set the x positions of the bars
r1 = np.arange(len(fixed))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bars
plt.bar(r1, fixed, color='b', width=bar_width, edgecolor='grey', label='Fixed')
plt.bar(r2, heuristic, color='g', width=bar_width, edgecolor='grey', label='Heuristic')
plt.bar(r3, ml_model, color='r', width=bar_width, edgecolor='grey', label='RL-Model')

# Add labels, title, and legend
plt.xlabel("Traffic light situation")
plt.ylabel("Average Car Travel Time (seconds)")
plt.title("Fig. 2 - Traffic Control Models Comparison")
plt.xticks([r + bar_width for r in range(len(fixed))], labels)
plt.legend(loc='lower right')

# Remove right and top axis spines
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

# Set a gray background
plt.gca().set_facecolor('lightgrey')

# Add a grid
plt.grid(axis='y', color='white', linestyle='-', linewidth=0.5)

# Set a less abrasive color for the bars
plt.gca().set_prop_cycle('color', ['#6EB5FF', '#9AFF9A', '#FF9999'])
# Show the plot
plt.show()
```
